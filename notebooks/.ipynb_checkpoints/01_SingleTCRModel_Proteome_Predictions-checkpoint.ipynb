```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# load in proteome
netmhc_proteome = pd.read_csv('../data/netmhc_WBSB_9mers_clean.csv')

# load in scores
for tcr_i in ['19.2']:
    netmhc_proteome[f'{tcr_i}_Scores'] = sigmoid(np.load(f'../weights/single_tcr/lightning_logs/{tcr_i}_finetune_esm/prediction/y_pred_proteome.npy').reshape(-1))
```

```python
# see score distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=False)

# Full histogram
ax1.hist(netmhc_proteome['19.2_Scores'], bins=50, color='gray', edgecolor='black')
ax1.set_xlabel('Scores', fontsize=16)
ax1.set_ylabel('Frequency', fontsize=16)
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.set_title('Full Distribution', fontsize=14)

# Zoomed-in histogram
counts, bins, _ = ax2.hist(netmhc_proteome['19.2_Scores'], bins=50, color='gray', edgecolor='black')
ax2.set_xlim(0.95, 1)
ax2.set_ylim(0, max(counts[(bins[:-1] >= 0.9) & (bins[1:] <= 1)]) * 1.2)
ax2.set_xlabel('Scores', fontsize=16)
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.set_title('Zoomed In on Scores 0.9–1.0', fontsize=14)

plt.tight_layout()
plt.show()
```
